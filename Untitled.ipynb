```python
# load libraries and set plot parameters
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'png')
plt.rcParams['savefig.dpi'] = 75

plt.rcParams['figure.autolayout'] = False
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 14

plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = "serif"
plt.rcParams['font.serif'] = "cm"

import scipy.constants as cte
from scipy.integrate import simps
from scipy.signal import gaussian

au_l   = cte.value('atomic unit of length')
au_t   = cte.value('atomic unit of time')
au_e   = cte.value('atomic unit of energy')
ev     = cte.value('electron volt')
au2ang = au_l / 1e-10
au2ev  = au_e / ev
```

```python
def segregation(x0, R, n, N=None):
    if N:
        return x0*(1-R**N)*R**(n-N)
    return x0*(1-R**n)

def concentration(x0, R, ml):
    nsl = 3 * ml # number of surrounding layers (each side)
    tl = ml + 2 * nsl  # total layers
    # bulk before well
    x = [0] * nsl
    # well
    for n in range(1, ml+1):
        x.append(segregation(x0, R, n))
    # barrier after well
    for n in range(1, nsl+1):
        x.append(segregation(x0, R, n+ml, ml))
    return np.arange(1, len(x)+1), np.array(x)
```

```python
ml = 12 # mono layers
x0 = 0.11
R = 0.85
layers, in_x = concentration(x0, R, ml)
plt.scatter(layers, in_x)
plt.plot()
```

```python
gap_77 = np.vectorize(lambda x: 1.508-1.47*x+0.375*x**2)(in_x)

a0   = np.vectorize(lambda x:     6.0584 *x + (1-x)*  5.6533)
ac   = np.vectorize(lambda x:    -5.08   *x - (1-x)*  7.17  )
av   = np.vectorize(lambda x:     1.00   *x + (1-x)*  1.16  )
b    = np.vectorize(lambda x:    -1.8    *x - (1-x)*  1.7   )
c11  = np.vectorize(lambda x:     8.33   *x + (1-x)* 12.23  )
c12  = np.vectorize(lambda x:     4.53   *x + (1-x)*  5.71  )
me   = np.vectorize(lambda x:     0.023  *x + (1-x)*  0.067 )
mhhz = np.vectorize(lambda x:     0.263  *x + (1-x)*  0.333 )

plt.scatter(layers, gap_77)
plt.show()
```

```python
exx = eyy = (a0[0]-a0)/a0
ezz       = -2.0*c12*exx/c11
pe        = -av*(exx+exx+ezz)
qe        = -(b/2.0)*(exx+eyy-2.0*ezz)

gap_77_strain = gap_77 + ac*(exx+eyy+ezz)+pe+qe
gap_77_c = 0.7 * gap_77_strain
gap_77_v = 0.3 * gap_77_strain

plt.scatter(layers, gap_77_c)
plt.scatter(layers, -gap_77_v)
plt.show()
```
